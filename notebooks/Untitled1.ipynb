```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Ha2eV=27.211386245988


df=pd.read_csv("/home/yoshida/calc/fehmticn2/20251231_trj_lc_s0tomecp/20260304_orca_s0-mecp_tddft/data.csv")

td_ene=np.array(df.iloc[:,1:13])

s0_ene=np.array(df.iloc[:,13])
s0_ene=np.repeat(s0_ene[:, None], 12, axis=1)
td_ene=td_ene+s0_ene

energy = np.array(df.iloc[:,13])[:, None]
energy = np.hstack((energy, td_ene))

energy = energy - np.min(energy)
energy = energy*Ha2eV
print(energy.T)
```

```python
x = np.arange(energy.shape[0])*4 # 列番号


for i in range(10):
    plt.plot(x, energy.T[i], label=f"{i+1}")

plt.xlabel("x (column index)")
plt.ylabel("value")
plt.legend()
plt.show()
```
